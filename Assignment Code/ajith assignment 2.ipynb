### **Load Data (Watson Data & Sales Data)**

```python
df = pd.read_csv('D:\PYTHON\DATA SCIENCE\DATA\WA_Fn-UseC_-Marketing-Customer-Value-Analysis (1).csv')

sales= pd.read_csv('D:\PYTHON\DATA SCIENCE\DATA\sales.csv')
df
sales


```

###**1. Add Engaged Column**

```python
df['engaged'] = df['Response'].apply(lambda x: 1 if x == 'Yes' else 0)
df
```

### **2. Calculate Engagement Rate By Sales Channel**

```python
engagement_by_sales_channel_df =df.groupby('Sales Channel')['engaged'].sum()/df.groupby('Sales Channel')['engaged'].count()*100

engagement_by_sales_channel_df.columns = ['Not Engaged', 'Engaged']
engagement_by_sales_channel_df
```

### **3.Visualize Engaged Rate according to Sales Channel through Bar/Pie Chart**

```python
engagement_by_sales_channel_df.plot(
    kind='pie',
    figsize=(8, 8),  # Set the figure size
    autopct='%1.1f%%',  # Display percentages on the chart
    startangle=140,  # Start angle to make the chart more readable
    colors=['skyblue', 'orange'],  # Custom colors for slices
    legend=False  # Disable legend, since it's already in the chart
)
plt.show()
```

The **Agent** channel has the most engagement, showing that customers prefer personal interaction.  
Digital channels like **Web** and **Call Center** are being used but could be improved to increase engagement.  
The **Branch** channel is still important, showing that traditional methods are valued alongside digital options.
